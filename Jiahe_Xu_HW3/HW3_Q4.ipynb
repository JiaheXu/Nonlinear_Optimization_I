## part A

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
np.random.seed(0)
```

```python
n = 1000
m = 100

T = 100

A = np.random.normal(0,1,size = (m,n))
b = np.random.normal(0,1,size = m)
```

```python
np.mean(A)
```

```python
np.std(A)
```

```python
np.mean(b)
```

```python
np.std(b)
```

```python
ATA = A.T.dot(A)
gamma = 2

eigenvalue, eigenvector = np.linalg.eig(ATA)
lambda_max = np.max(eigenvalue)
lambda_min = np.min(eigenvalue)
print("Lipschitz gradient constant is: ",lambda_max)
print("Strong convexity constant is: ",lambda_min)
```

```python
def get_subgradient(x):
    #gradient = np.subtract(np.matmul(ATA,x),np.matmul(A.T,b))
    subgradient = np.matmul(ATA,x) - np.matmul(A.T,b)
    subgradient[x>0]+= gamma
    subgradient[x<=0]-= gamma
    return subgradient
```

## part B

```python
step_size=1/lambda_max
log_1 = []
obj_value = []
max_iters = T
x_k = np.zeros(n)

#kn1 means k+1
x_kn1 = np.zeros(n)
```

```python
for i in range(1,T+1):
    x_k = x_kn1
    subgradient = get_subgradient(x_k)
    log_1.append( np.linalg.norm(subgradient,2) )
    
    x_kn1 = x_k - step_size*subgradient
    cost = 1/2*( np.linalg.norm(A@x_kn1 - b ,2)**2) + gamma*np.linalg.norm(x_kn1 ,1)
    obj_value.append(cost)
    print("iteration: ", i, "\nobjective value is: ",cost,'\n')
```

```python
num = len(np.where(x_kn1 == 0)[0])
print("number of zeros: ",num)
```

## part C

```python
def get_proximal(x , alpha ):
    proximal = x
    proximal[ (-alpha<x) & (x<alpha)] = 0
    proximal[ (x >= alpha) ] -= alpha
    proximal[ (x <= -alpha) ] += alpha
    return proximal
```

```python
step_size=1/lambda_max
log_1 = []
obj_value = []
max_iters = T
x_k = np.zeros(n)

#kn1 means k+1
x_kn1 = np.zeros(n)
x = np.zeros(n)
```

```python
for i in range(1,T+1):
    x_k = x_kn1
    gradient = np.matmul(ATA,x_k) - np.matmul(A.T,b)
    x = x_k - step_size*gradient;
    
    proximal = get_proximal(x , step_size*gamma)
    x_kn1 = proximal
    
    cost = 1/2*( np.linalg.norm(A@x_kn1 - b ,2)**2) + gamma*np.linalg.norm(x_kn1 ,1)
    obj_value.append(cost)
    print("iteration: ", i, "\nobjective value is: ",cost,'\n')
```
