```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lambda0 = 1
x0 = [1/5 ,-1/5 ,4/5]
x0 = np.array(x0).reshape(3,1)
A = np.array([4,2,1,2,3,0,1,0,1])
A = A.reshape(3,3)
```

```python
eigvalues,eigvectors= np.linalg.eig(A)
```

```python
def F(x,Lambda):
    x = x.reshape(3,1)
    F = ( A - Lambda*np.eye(3) )@x
    #F = np.vstack(F, x.T@x -1 )
    F = np.concatenate( (F, x.T@x -1 ) )
    return F
```

```python
def grad_F(x,Lambda):
    x = x.reshape(3)
    grad = [  4 - Lambda, 2, 1,-x[0] ,
              2, 3 - Lambda,0,-x[1] ,
              1, 0, 1 - Lambda, -x[2] ,
              2*x[0], 2*x[1], 2*x[2],0 ]
    grad = np.array(grad,dtype = float)
    grad = grad.reshape(4,4)
    return grad
```

```python
def grad_F_inv(x,Lambda):
    grad = grad_F(x,Lambda)
    grad_inv = np.linalg.inv(grad)
    return grad_inv 
```

# y is the combination of x and lambda

```python
iter_max = 100
y_list = []
y_now =[]
y_next =np.array( [1/5 ,-1/5 ,4/5,1]).reshape(-1,1)
# y is the combination of x and lambda

Lambda = lambda0
x_now = x0
x_next = []

steps = 0

while np.abs( Lambda - eigvalues[1]) > 1e-3 :
    
    y_now = y_next
    y_next = y_now - grad_F_inv(x_now,Lambda)@F(x_now,Lambda)
    x_now = y_next[0:3 , :]
    Lambda = y_next[3]
    
    y_list.append(y_next)
    steps = steps+1
    
print("last y: \n",y_list[-1])
print("iteration steps: ", steps)
```

```python
def h(y):
    x = y[0:3]
    Lambda = y[3]
    val = 1/2* ( np.linalg.norm(F(x,Lambda)) **2)
    
    return val
```

```python
def grad_h(x,Lambda):
    
    grad = grad_F(x,Lambda).T @ F(x,Lambda)
    
    return grad
```
